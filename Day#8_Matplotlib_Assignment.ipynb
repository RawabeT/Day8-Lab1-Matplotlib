```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
lst = [1,2,3,4,5,6,7,8]
df = pd.DataFrame(np.random.randint(0,100,size=(10, 4)), columns=list('ABCD'))


plt.figure(figsize=[20,5])
plt.suptitle('Subplots',fontsize = 16)

plt.subplot(1,2,1)
plt.plot(lst);

plt.subplot(1,2,2)
plt.plot(df);
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,5), labels =['Sun','Mon','Tue','Wed'], explode = [.0,.1,.3,.8]);
plt.title('Days', fontsize = 20);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
scatter = plt.scatter(np.arange(1,5), np.arange(1,5), alpha = 0.4, s = 50);


#Q4 Ans:
plt.title('Dotes',loc='left', c = 'r')
plt.xlabel('row', fontsize =16)
plt.ylabel('column', fontsize =16)
```
